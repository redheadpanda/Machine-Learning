**Imbalanced Dataset:**

A dataset with an unequal class distribution.
Like if there are 80 percent of class 1 and 20 percent of class 2

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data['Class'].value_counts()
```

```python
# separate data of both classes
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
legit.shape
#fraud.shape
```

```python
legit_sample = legit.sample(n=213)
```

So now we have legit - 213 as well as fraud - 213.
Now all we have to do is to concatenate both data

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```
